```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv('/home/afsal/Desktop/vs_code/afsal_python/cs50_ai/ml/code_basics/week_3/test_scores.csv')
df.head()
```

```python
x = df['math']
y = df['cs']
plt.scatter(x, y)
```

```python
def gradient_descent(x, y):
    iterations = 10000
    learning_rate = 0.0002
    n = len(x)
    theta_0 = theta_1 = prev = 0
    
    
    for i in range(iterations):
        y_pred = (theta_1*x) + theta_0
        coef_grad = -sum((y-y_pred)*x)/n
        int_grad = -sum(y-y_pred)/n
        theta_0 = theta_0 - (learning_rate * int_grad)
        theta_1 = theta_1 - (learning_rate * coef_grad)
        cost = sum([i**2 for i in (y-y_pred)])/n
        
        if i == 1:
            print(f'initail cost: {cost}')
            
        if math.isclose(cost, prev):
            break
        
        prev = cost
        
        
    print(f'coef = {theta_1}, intercept = {theta_0}, cost = {cost}, iteration = {i}')
    return theta_1, theta_0
    
def abline(slope, intercept):
    """Plot a line from slope and intercept"""
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--')

x = df['math']
y = df['cs']
plt.scatter(x, y)



coeff, intercept = gradient_descent(x, y)
abline(coeff, intercept)

```

```python
class line_model:
    def __init__(self, m, b) -> None:
        self.m = m
        self.b = b 
    
    def predict(self, x):
        return self.m*x + self.b
    
model = line_model(coeff, intercept)
model.predict(23)
```
