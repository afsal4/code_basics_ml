```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def gradient_descent(x, y):
    print(x, y)
    iterations = 500
    learning_rate = 0.005
    b = m = 0
    n = len(x)
    for i in range(iterations):
        predicted_val = (m*x) + b
        vary = (y - predicted_val)
    
        md = (-1/n) * sum(x*vary)
        bd = (-1/n) * sum(vary) 
        m = m - (learning_rate * md)
        b = b - (learning_rate * bd)
    return m, b
```

```python
def plot_line(coef, intercept, x):

    y = []
    for i in x:
        y.append(coef*(i)+intercept)
    y = np.asarray(y)
    plt.plot(x, y)
```

```python

x = np.array([1, 2, 3, 5])
y = np.array([2, 5, 7, 10])
l, o = gradient_descent(x, y)
plt.scatter(x, y)

print(l, o)
plot_line(l, o, x)
```
