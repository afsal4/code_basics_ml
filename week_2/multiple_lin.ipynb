```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('homeprices.csv')
df.head()
```

```python
median_no = int(df['bedrooms'].median())
```

```python
# filling empty value
df.bedrooms = df.bedrooms.fillna(median_no)
```

```python
model = LinearRegression()
x = df.iloc[:, :-1]
y = df['price']

model.fit(x, y)
predicted_vals = model.predict(x)

plt.plot(x, y, color='black')
plt.plot(x, predicted_vals, color='blue')
```
